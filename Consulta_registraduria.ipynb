```python
import csv
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

def parametros():    
    global registro
    registro = 1
    global options
    options = webdriver.ChromeOptions()
    options.add_argument('__incognito')
    # Crear una instancia del navegador
    global driver
    driver = webdriver.Chrome(options=options)
    # Cargar la p치gina web
    driver.get('https://defunciones.registraduria.gov.co')
    # Esperar a que la p치gina cargue
    time.sleep(5)
    
def abrir_txt():    
     with open('cedulas.txt', 'r') as archivo_entrada:
         lector_csv = csv.reader(archivo_entrada)
         with open('resultados2.csv', 'w', newline='') as archivo_salida:
             global escritor_csv
             escritor_csv = csv.writer(archivo_salida)
             for cedula in lector_csv:  
                 #print(cedula)
                 try:
                     consultar(cedula)
                 
                 except Exception as e:
                     #print(chr(27)+"[1;30m"+"Tipo Exepcion : ")                     
                     #print(type(e))                           # the exception type
                     #print("Tipo Argumentos de exepcion : ")
                     #print(e.args)                            # arguments stored in .args
                     print(chr(27)+"[1;30m"+"Exepcion : ")
                     print(e)                                 # __str__ allows args to be printed directly,
                                                              # but may be overridden in exception subclasses
                     reintentar(cedula)
                 """except NoSuchWindowException as i:
                     print(chr(27)+"[1;30m"+"Tipo Exepcion : ")                     
                     print(type(i))                           # the exception type
                     print("Tipo Argumentos de exepcion : ")
                     print(i.args)                            # arguments stored in .args
                     print("Exepcion : ")
                     print(i)                                 # __str__ allows args to be printed directly,
                                                              # but may be overridden in exception subclasses
                     reintentar(cedula)"""                                    

def consultar(cedula):
    #print("Registro : ",registro)
    #print(chr(27)+"[1;30m"+"Cedula a consultar : ", cedula) 
    #print("Cedula a consultar : ", cedula)
    cuadro_texto = driver.find_element(By.XPATH, '//*[@id="nuip"]')
    cuadro_texto.send_keys(cedula)
    cuadro_texto.send_keys(Keys.RETURN)
    # Esperar a que la p치gina cargue
    time.sleep(1)
    elemento_mensaje = driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div/div/div/div/p')
    mensaje = elemento_mensaje.text
    #print("Menssaje : ",mensaje)
    # Escribir el resultado en el archivo de salida
    escritor_csv.writerow([cedula, mensaje])
    #registro=registro+1

def iniciar():
    print(chr(27)+"[3;32m"+"SE INICIO LA CONSULTA")
    
def cerrar():
    print(chr(27)+"[3;44m"+"SE TERMINO LA CONSULTA")
    driver.quit()
    
def reintentar(cedula):
    print(chr(27)+"[1;31m"+"--ERROR EN LA CONSULTA--- Cedula :", cedula)                      
    driver.quit()
    registro=1
    parametros()
    print(chr(27)+"[1;33m"+"--INTENTANDO CONSULTAR LA Cedula :", cedula)                      
    consultar(cedula)   
    

iniciar()
parametros()
abrir_txt()
cerrar()
```
